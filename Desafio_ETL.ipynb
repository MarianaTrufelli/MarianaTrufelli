```
import pandas as pd
import json
import random

# List files recursively to find 'leitores.csv'
!find /content/ -name '*leitores.csv*'

# EXTRACT
try:
  df = pd.read_csv('/content/"leitores.csv"', encoding='utf-8')
  print("Dados extraídos com sucesso!")
  print(df.head())
except FileNotFoundError:
  raise FileNotFoundError("Arquivo 'leitores.csv' não encontrado.")

# Verificação de colunas obrigatórias
required_columns = {'name', 'favorite_genre'}
if not required_columns.issubset(df.columns):
  raise ValueError(f"O CSV deve conter as colunas: {required_columns}")

# Remove linhas com valores nulos nas colunas críticas
df = df.dropna(subset=required_columns)

# TRANSFORM
recomendacoes = {
    'Ficção Científica': [
        'Duna (Frank Herbert)',
        'Neuromancer (William Gibson)',
        'O Guia do Mochileiro das Galáxias (Douglas Adams)'
    ],'Romance': [
        'Orgulho e Preconceito (Jane Austen)',
        'Como Eu Era Antes de Você (Jojo Moyes)',
        'A Culpa é das Estrelas (John Green)'
    ],
    'Fantasia': [
        'O Senhor dos Anéis (J. R. R. Tolkien)',
        'Harry Potter (J. K. Rowling)',
        'O Nome do Vento (Patrick Rothfuss)'
    ],
    'Mistério': [
        'O Código Da Vinci (Dan Brown)',
        'Garota Exemplar (Gilliam Flynn)',
        'Sherlock Holmes (Arthur Conan Doyle)'
    ],
    'Não Ficção': [
        'Sapiens (Yuval Noah Harari)',
        'Como Fazer Amigos e Influenciar Pessoas (Dale Carnegie)',
        'Armas, Dermes e Aço (Jared Diamond)'
    ]
}

def gerar_mensagem(row):
  nome = row['name']
  genero = row['favorite_genre']

  if genero in recomendacoes:
    livro = random.choice(recomendacoes[genero])
    return (
        f"Oi, {nome}! Baseado no seu interesse em {genero},"
        f"recomendamos o livro '{livro}'. Ótima leitura!"
    )

  return (
      f"Olá, {nome}! Que tal explorar novos gêneros?"
      f"Recomendamos começar por Ficção Científica!"
  )

  # Aplica a transformação
  df['news'] = df.apply(gerar_mensagem, axis=1)

  print("\nDados transformados:")
  print(df.head())

  # LOAD
  df.to_csv(
      'users_transformed.csv',
      index=False,
      encoding='utf-8'
  )

  with open('users_transformed.json', 'w', encoding='utf-8') as f:
    json.dump(
        df.to_dict(orient='records'),
        f,
        ensure_ascii=False,
        indent=4
    )

    print("\nDados carregados com sucesso!")
    print("Arquivos gerados:")
    print("- users_transformed.csv")
    print("- users_transformed.json")
```
